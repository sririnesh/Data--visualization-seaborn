# importing the libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# importing the dataset

```python
dataset=pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
dataset
```

```python
# number of rows and columns
print("The Number of Row:",dataset.shape[0])
print("The Number of columns:",dataset.shape[1])
```

# univariate Analysis

```python
import seaborn as sb
sb.distplot(dataset["10th %"],hist=True,kde_kws={"color":"red"})
```

# Bivariate Analysis: Distribution plot

```python
sb.jointplot(x= "10th %",y= "salary",data=dataset)
plt.show()
```

```python
sb.jointplot(x= "10th %",y= "salary",data=dataset,kind="hex")
plt.show()
```

```python
sb.jointplot(x= "10th %",y= "salary",data=dataset,kind="kde")
plt.show()
```

```python
#pair plot
```

```python
sb.pairplot(dataset,hue="Gender",diag_kind="kde",kind="scatter",palette="husl")
plt.show()
```

# For categorical plot

```python
sb.stripplot(x= "SSC Board",y= "10th %",data=dataset)
plt.show()
```

```python
sb.stripplot(x= "Gender",y= "10th %",data=dataset,jitter=True)
plt.show()
```

```python
sb.swarmplot(x= "SSC Board",y= "10th %",data=dataset)
plt.show()
```

# Distribution observation

```python
sb.violinplot(x= "Gender",y= "10th %",data=dataset)
plt.show()
```
